```
def process_image(image_path1,image_path2):
  img1 = cv2.imread(image_path1, 1)
  img1=cv2.resize(img1,(160,160))
  img1 = img1[...,::-1]
  img1 = np.around(np.transpose(img1, (2,0,1))/255.0, decimals=12)
  x_train1 = np.array([img1])
  x_train1=np.rollaxis(x_train1, 1, 4)
  l1=model.predict_on_batch(x_train1)

  img2 = cv2.imread(image_path2, 1)
  img2=cv2.resize(img2,(160,160))
  img2 = img2[...,::-1]
  img2 = np.around(np.transpose(img2, (2,0,1))/255.0, decimals=12)
  x_train2 = np.array([img2])
  x_train2=np.rollaxis(x_train2, 1, 4)
  l2=model.predict_on_batch(x_train2)

  dist = np.linalg.norm(l1-l2)
  print(dist)
  if(dist<1):
    print("yes")
  else:
    print("no")

```
